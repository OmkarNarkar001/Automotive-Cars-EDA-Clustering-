# Replacing NULL values

```python
# Replace NaN value in num-of-door column based on car length
def determine_num_doors(length):
    return "Four" if length > 170 else "Two"

df["num-of-doors"].fillna(df["length"].apply(determine_num_doors), inplace=True)

# Replace NaN values in bore & stroke column based on fuel-type, compression-ratio, body-style
filtered_data = df[(df["fuel-type"] == "gas") & 
                   (df["compression-ratio"] == 9.4) & 
                   (df["body-style"] == "hatchback")
                  ]

for column in ["bore", "stroke"]:
    df[column].fillna(filtered_data[column].mode()[0], inplace=True)

# Replace NaN values in horsepower & peak-rpm column based on other features of the car
filtered = df[(df["fuel-type"] == "gas") & 
              (df["num-of-cylinders"] == "four") & 
              (df["fuel-system"]== "mpfi") & 
              (df["engine-type"] == "ohc") & 
              (df["compression-ratio"] == 8.70)
             ]

for column in ["horsepower", "peak-rpm"]:
    mode_value = filtered[column].mode()[0]
    df[column].fillna(mode_value, inplace=True)

# Convert "price" column to numeric
df["price"] = pd.to_numeric(df["price"], errors='coerce')

# Fill NaN values in the "price" column based on other features of the car
df["price"].fillna(df[(df["fuel-type"] == "gas") & (df["body-style"] == "sedan") & (df["num-of-doors"] == "two") & (pd.to_numeric(df["horsepower"], errors='coerce') < 90)]["price"].mean(), inplace=True)

# Replace NaN values in the "price" column for Audi cars
mean_price_audi = df[df["make"] == "audi"]["price"].mean()
df.loc[df["make"] == "audi", "price"].fillna(mean_price_audi, inplace=True)

# Replace NaN values in the "price" column for Porsche cars
mean_price_porsche = df[(df["make"] == "porsche") & (df["body-style"] == "hatchback")]["price"].mean()
df.loc[df["make"] == "porsche", "price"].fillna(mean_price_porsche, inplace=True)
```

### OneHotEncoding

```python
df_duplicated = df.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Assuming df_duplicated is defined and contains the required columns
df_encoder_columns = ["drive-wheels","engine-location","make","fuel-type", "aspiration", "num-of-doors", "body-style",
                      "engine-type", "num-of-cylinders", "fuel-system"]

encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(df_duplicated[df_encoder_columns])

# Create a DataFrame with the encoded data
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(df_encoder_columns))

# Concatenate the encoded DataFrame with the original DataFrame
encoded_data = pd.concat([df_duplicated.drop(columns=df_encoder_columns), encoded_df], axis=1)
```

```python
encoded_data.head()
```

## KNN to replace Null values

```python
# Replacing Null values of Normalized_losses columns through KNN Imputer

from sklearn.impute import KNNImputer

# Identify features with missing values
missing_cols = encoded_data.columns[encoded_data.isnull().any()].tolist()

# Create a copy of the DataFrame
df_imputed = encoded_data.copy()

# Initialize KNN imputer
imputer = KNNImputer(n_neighbors=5)  

# Impute missing values
imputed_values = imputer.fit_transform(df_imputed)

# Replace missing values in the original DataFrame
df_imputed[missing_cols] = imputed_values[:, encoded_data.columns.isin(missing_cols)]
```
